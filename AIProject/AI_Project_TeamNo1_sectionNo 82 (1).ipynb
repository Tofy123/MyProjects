```python
for column in columnns:
    top10 = health.nlargest(5,column)
    plt.figure(figsize=(10,4))
    plt.bar(top10['City'], top10[column],color='green',alpha=0.9)
    plt.ylabel(column, fontsize = 10)
    plt.xlabel("Country", fontsize = 10)
    plt.title("Top 5 countries per y", fontsize = 10)
```

```python
for column in columnns:
    top10 = health.nsmallest(5,column)
    plt.figure(figsize=(10,4))
    plt.bar(top10['City'], top10[column],color='black',alpha=0.5)
    plt.ylabel(column, fontsize = 10)
    plt.xlabel("Country", fontsize = 10)
    plt.title("Least 5 countries per y", fontsize = 10)
```

```python
data=health[['Obesity levels(Country)', 'Life expectancy(years) (Country)']] # we will consider two features for the clustering 
data.head()
```

```python
scaler = StandardScaler().fit(data.values) # fitting thr standizer 
features = scaler.transform(data.values) # transforming the data 
scaled = pd.DataFrame(features, columns = data.columns) # creating a data frame with the scaled data
scaled.head()
```

```python
from sklearn.cluster import KMeans

clusters_range=[2,3,4,5,6,7,8,9,10] # to test diffrent number of clusters 
centroid=[]
```

```python
for c in clusters_range:
    kmeans=KMeans(n_clusters=c, random_state=0).fit(scaled)
    centroid.append(kmeans.inertia_)

plt.figure(figsize=(7,7))
plt.plot(clusters_range, centroid, marker='o')
```

```python
#Using 6 clusters
kmeans=KMeans(n_clusters=6, random_state=17) 
kmeans.fit(scaled)
y_pred=kmeans.fit_predict(scaled)
print(y_pred)
```

```python
plt.scatter(scaled['Obesity levels(Country)'], scaled['Life expectancy(years) (Country)'], c=y_pred)
plt.title("KMean Clustering with K equals to 6")
plt.show()
```

```python
#Hierarchical Clustring
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='ward')
y_pred = cluster.fit_predict(scaled)
print(y_pred)

plt.scatter(scaled['Obesity levels(Country)'], scaled['Life expectancy(years) (Country)'], c=y_pred, cmap='rainbow')
plt.title("Hierarchical Clustring with K equals to 6")
plt.show()
```

```python
#Gaussian Mixture model
import numpy as np
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=6, random_state=14).fit(scaled)
gm.means_
y_pred = gm.predict(scaled)

plt.scatter(scaled['Obesity levels(Country)'], scaled['Life expectancy(years) (Country)'], c=y_pred)
plt.title("Gaussian Mixture Model Clustering with K equals to 6")
plt.show()
```
